```python
# 시각화를 위해 한글 깨짐 현상 방지를 위해 폰트 설치 및 사용 (설치 후 반드시 메뉴 [런타임] -> [런타임 다시 시작] 하기)
%matplotlib inline  

import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm  

!apt-get update -qq
!apt-get install fonts-nanum* -qq

path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf' 
font_name = fm.FontProperties(fname=path, size=10).get_name()
print(font_name)
plt.rc('font', family=font_name)

fm._rebuild()
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 한국어 명사 추출을 통해 워드 카운팅 및 워드클라우드 그리기 위해 필요한 NLP 모듈인 KONLPY 설치
!pip install konlpy
```

```python
# 한국어 자연어처리 중 명사 추출을 위해 필요한 모듈 importing하기
from konlpy.tag import Okt
okt = Okt()
```

```python
# 명사추출하기
rs = okt.nouns(txt)
```

```python
# 추출한 명사에 대한 워드카운팅 및 그 결과를 data_dict1이라는 딕셔너리에 저장하기
data_dict1 = {}
for i in rs:
  if i in data_dict1:
    data_dict1[i] += 1
  else:
    data_dict1[i] = 1
data_dict1
```

```python
# 워드 카운팅 시각화를 위해 사전을 데이터프레임으로 만들고, 막대 그래프 그리기
# 금칙어 (stopwords)를 지정하고 삭제할 수 있음
import pandas as pd

stopwords = ['등', '것이다']
for i in stopwords:
  try: 
    del data_dict1[i]
  except:
    pass

df = pd.DataFrame(data_dict1, index=[0])
dff = df.T
dff.columns= ['빈도']
fig, ax = plt.subplots(figsize=(25,10))
plt.rcParams['font.family'] = 'NanumBarunGothic'

dff[dff['빈도'] >=3].plot(kind='bar', ax=ax)
plt.legend('F')
plt.xticks(rotation=60)
plt.show()
```

```python
# 단어의 등장 빈도로 워드 클라우드 찍기 위해서 단어들을 다시 하나의 텍스트로 만들기
list_wordcld = ''
for i in data_dict1:
    for j in range(data_dict1[i]):
        list_wordcld += f'{i} '
```

```python
# 워드클라우드를 그리기 위해 필요한 모듈 importing 및 워드클라우드 클래스 생성, 끝으로 시각화하기
from wordcloud import WordCloud    
word_string = list_wordcld
STOPWORDS = ['것']
path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
wordcloud = WordCloud(font_path=path,
                          background_color="white",
                          width=1200, height=1000,
                          collocations = False,
                          colormap='prism',
                          stopwords=STOPWORDS
                         ).generate(word_string)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud)
plt.xticks([])
plt.yticks([])
plt.show()
```
